```python
import pandas as pd
import numpy as np
migration = pd.read_csv('./data/Wave 3 Server List.csv')
migration = migration.sort_values(by = ['App Name'])
snow = pd.read_csv('./data/ServiceNow.csv')
snow = snow.sort_values(by = ['name'])
```

```python
migration.head(5)
```

```python
snow['App Name'] = snow['name'] 
snow.head(5)
```

```python
output = migration.copy(deep='True')

for index, row in snow.iterrows():
    app_name = row['name']
    bc = row['business_criticality']
    data = output.loc[(output['App Name'] == app_name)]
    for index in data.index:
        output.at[index, 'Business Criticality'] = bc
```

```python
diff = snow.join(output.set_index('App Name'), on='App Name')

for index in diff.index:
    if diff.iloc[index]['name'] is not diff.iloc[index]['App Name']:
        print(diff.iloc[index]['name', 'business_criticality', 'business_criticality'])
    if diff.iloc[index]['business_criticality'] is not diff.iloc[index]['business_criticality']:
        print(diff.iloc[index]['name', 'business_criticality', 'business_criticality'])
        
print(diff[['name', 'business_criticality', 'Business Criticality']])
```
